## 作業目標

根據範例：

1. 取出今天所有的新聞
2. 取出現在時間兩小時內的新聞
3. 根據範例，取出三天前下午三點到五點的新聞

## 1. 取出今天所有的新聞

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time

browser = webdriver.Chrome(executable_path='chromedriver')

browser.get("https://www.ettoday.net/news/news-list.htm")

today = time.strftime("%Y/%m/%d", time.localtime())

while True:
    time.sleep(0.2)
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, "html5lib")
    d = soup.find(class_="part_list_2").find_all('h3')
    if d[-1].find(class_="date").text[0:10] != today:
        break
    
```

```python
for d in soup.find(class_="part_list_2").find_all('h3'):
    if d.find(class_="date").text[0:10] == today:
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
    
```

## 2. 取出現在時間兩小時內的新聞

```python
import datetime
time_1 = datetime.datetime.now() - datetime.timedelta(hours=2)
#print(time_1)
for d in soup.find(class_="part_list_2").find_all('h3'):
    if datetime.datetime.strptime(d.find(class_='date').text, '%Y/%m/%d %H:%M') > time_1:
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

```python
browser.quit()
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
browser = webdriver.Chrome(executable_path='chromedriver')
browser.get("https://www.ettoday.net/news/news-list.htm")
from selenium.webdriver.support.ui import Select
time_1 = (datetime.datetime.now() - datetime.timedelta(days=3)).strftime('%Y/%m/%d')

pm3 = datetime.datetime.strptime(time_1 + ' 15:00', '%Y/%m/%d %H:%M')
pm5 = datetime.datetime.strptime(time_1 + ' 17:00', '%Y/%m/%d %H:%M')

print(pm3,pm5)
```

```python
selY = Select(browser.find_element_by_id("selY"))
selY.select_by_value(datetime.datetime.strptime(time_1, '%Y/%m/%d').strftime('%Y'))
selM = Select(browser.find_element_by_id("selM"))
selM.select_by_value(str(int(datetime.datetime.strptime(time_1, '%Y/%m/%d').strftime('%m'))))
selD = Select(browser.find_element_by_id("selD"))
selD.select_by_value(datetime.datetime.strptime(time_1, '%Y/%m/%d').strftime('%d'))
browser.find_element_by_id('button').click()
```

```python

while True:
    time.sleep(0.2)
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, "html5lib")
    d = soup.find(class_="part_list_2").find_all('h3')
    if datetime.datetime.strptime(d[-1].find(class_="date").text, '%Y/%m/%d %H:%M') < pm3 :
        break
```

```python
for d in soup.find(class_="part_list_2").find_all('h3'):
    if datetime.datetime.strptime(d.find(class_='date').text, '%Y/%m/%d %H:%M') >= pm3 and datetime.datetime.strptime(d.find(class_='date').text, '%Y/%m/%d %H:%M') < pm5:
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

```python
browser.quit()
```
