```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(r.text, 'html.parser')
table = soup.find('table', class_='TABLE_G')
print(table)
```

```python
d = {}
trs = table.find_all('tr' )
for tr in trs:
    tds = tr.find_all('td')
    if len(tds) == 5:
        item = tds[0].text
        #print(item)
        d[item] = {}
        d[item][tds[2].text] = tds[3].text
    else :
        try:
            #print(tds[0].text)
            #print(tds[1].text)
            d[item][tds[0].text] = tds[1].text
        except:
            pass
        
print(d)
```
